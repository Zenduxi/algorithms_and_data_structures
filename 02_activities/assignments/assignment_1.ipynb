
-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
# Your answer here
from typing import List

def first_duplicate(nums: List[int]) -> int:
    size = len(nums)
    keys = range(size)  # O(n)
    num_dict = dict(zip(keys, nums))  # O(n)
    sorted_num_list = sorted(num_dict.items(), key=lambda item: item[1]) # O(nlog(n))
    min_dup_key = None
    for i in range(size-1):  # O(n)
        if sorted_num_list[i][1] == sorted_num_list[i+1][1] and ( min_dup_key is None or min_dup_key > sorted_num_list[i][0]):
            min_dup_key = sorted_num_list[i][0]

    if min_dup_key is None:
        return(-1)
    else:
        return(num_dict[min_dup_key])
 

first_duplicate([99,53,28,100,100,100,50,50])
````


-   Explain the thinking to an alternative solution (no coding required, but a classmate reading this should be able to code it up based off your text)


````python
# Your answer here
# We can also use two for loops to solve the problem as shown below.
# The time complexity is O(n^2)
from typing import List

def first_duplicate_v2(nums: List[int]) -> int:
    size = len(nums)
    for i in range(size-1):
        if nums[i] in nums[(i+1):]:
            return nums[i]
    return(-1)
````

````python
# Test the time for each of the two aforementioned methods
import timeit
import random

a = list(range(10**4))
a.append(200)
a.append(400)
random.shuffle(a)

sort_time = timeit.timeit(lambda:first_duplicate(a), number = 1)
print(f'{sort_time:.6f}')

loop_time = timeit.timeit(lambda:first_duplicate_v2(a), number = 1)
print(f'{loop_time:.6f}')

````
