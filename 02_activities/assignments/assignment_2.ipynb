  # Question One: Check Duplicates in Tree

  Given the `root` of a binary tree, check whether it is contains a duplicate value. If a duplicate exists, return the duplicate value. If there are multiple duplicates, return the one with the closest distance to the root. If no duplicate exists, return -1.

#### Starter Code for Question 1

```python
# Definition for a binary tree node.
from collections import deque 

class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right


def is_duplicate(root: TreeNode) -> int:
    # Breadth First Search is used in this function. So that the first found duplicate is the closest one to the root. 
    # The set, visited, contains the distinct values of visited nodes.
    # If current_node.val is in the set, the first duplicate is found. Or else add the value to the set.
    # The children of the current node is added to node_to_visit.
    # At the beginning of the while loop, the first object in node_to_visit becomes the current_node.
    current_node = root
    visited = set([root.val]) 
    node_to_visit = deque([current_node.left, current_node.right])

    while node_to_visit:
        current_node = node_to_visit.popleft()
        if current_node.val in visited:
            return current_node.val
        
        visited.add(current_node.val)
        if current_node.left is not None:
            node_to_visit.append(current_node.left)
        if current_node.right is not None:
            node_to_visit.append(current_node.right) 
    return -1      


def construct_tree(node_list: list) -> TreeNode:
    # This function is used to convert the given list to a tree.
    # Breadth First Search traversal method is used based on the given examples.
    # The given list is put in a queue.
    # When visit a node, it is current_node. The first interger in the queue is popped and assgined to the current_node.val.
    # The 2nd and 3rd in the queue are poped and used to create the child nodes of the current node. The childen are added to node_to_visit.
    # At the beginning of the while loop, the first object in node_to_visit becomes the current_node.
    queue = deque(node_list)
    bfs = TreeNode(queue.popleft())
    node_to_visit = deque([bfs])
    
    while queue:
        current_node = node_to_visit.popleft()

        current_node.left = TreeNode(queue.popleft())
        node_to_visit.append(current_node.left)

        if queue:
            current_node.right = TreeNode(queue.popleft())
            node_to_visit.append(current_node.right)

    return bfs
        
            
        
```

```python
# root = [1, 2, 2, 3, 5, 6, 7]
# root = [1, 3, 5, 6, 7,6]
root = [1, 10, 2, 3, 10, 12, 12]
# root = [10,9,8,7]
bfs = construct_tree(root)
is_duplicate(bfs)
```

```python
# Your answer here
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:
    # go through each integer in the list and add non-zeroes to seperate list as they appear and update running counter to keep track of number of 0's. After processing list of integers add append counted number of 0's to non-zeroes list.
    processed_list = []
    zero_counter = 0
    for i in nums:
        if i == 0:
            zero_counter += 1
        else:
            processed_list.append(i)
    for j in range(zero_counter):
        processed_list.append(0)
    return processed_list
```
